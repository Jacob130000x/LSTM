```python
dfStock = pd.read_csv('APPL.csv') 
dfStock.shape
#Adj closing
dfStockClose = dfStock.filter(['Adj Close'])
#converting to numpy array
dataset = dfStockClose.values
lenData = len(dataset)


#LSTMPred(dataset1)



def LSTMPred(dataset):

  dataset1 = dataset[:2000]
  lenData1 = len(dataset1)
  RESULTS = []
  
  for i in range(500):

    
    dataRange = lenData1 + i
    dataset3 = dataset[0:dataRange, ]
    print('dataset', dataset3)
  

    #training rows
    constantPercentage = 0.8
    training_data_len = len(dataset3) - 1
    training_data_len
    #Scale the data
    #///////////////MIN MAX SCALER///////////
    # scaler = MinMaxScaler(feature_range=(0,1))
    # scaled_data = scaler.fit_transform(dataset3)
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(dataset3)

    #//////////////STANDARD SCALER////////
    # scaler = StandardScaler()
    # #scaled_data = scaler.fit(dataset3)
    # scaled_data = scaler.fit_transform(dataset3)
    # scaled_data = scaler.transform([[0, 10]])

   # scaled_data = preprocessing.normalize(dataset3)
   

    #create training dataset
    train_data = scaled_data[0:training_data_len, :]
    # split the data x y train
    x_train = []
    y_train = []

    # rangeOfData =len(train_data)
    for j in range(1, len(train_data)):
      x_train.append(train_data[j-1 : j, 0])
      y_train.append(train_data[j,0])
    
    # convert x y train to numpy array
    x_train, y_train = np.array(x_train), np.array(y_train)
  #Reshape the data

    x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
    x_train.shape

    model = Sequential()
    model.add(LSTM(100,return_sequences=True, input_shape = (x_train.shape[1], 1)))
    model.add(LSTM(100,return_sequences=False))
    
    model.add(Dense(100))
    model.add(Dense(1))

    model.compile(optimizer='SGD', loss = 'mean_squared_error')
    # model.compile(optimizer='adam', loss = 'mean_squared_error')
    model.fit(x_train, y_train, batch_size=1, epochs=1)
    #Testing set
    test_data =scaled_data[-1]
    print('test data before prediction', test_data)
    x_test = []
    y_test = dataset3[-1] #actual last value for prediction
    print('y_test before prediction', y_test)
    x_test.append(test_data)
    x_test = np.array(x_test)
    x_test = np.reshape(x_test, (1, 1, 1))
  #Predictions
    predictions = model.predict(x_test)
    predictionsNotScaled = scaler.inverse_transform(predictions)
  #RMSE
    rmse = np.sqrt(np.mean(predictionsNotScaled - y_test)**2)
    print(rmse)


    print('Prediction not scaled', predictionsNotScaled)
    print('actual result', y_test ) 
    print('x_test', x_test)
    #print('xtest_Inverse', xtest_Inverse)
    print('day before',dataset3[-2])
    print('today',dataset3[-1])

    upPred = 'j'
    downPred = 'h'
    upActual = 'g'
    downActual = 'z'

    if predictionsNotScaled > dataset3[-2]:
      upPred = 'x'
    
    if predictionsNotScaled < dataset3[-2]:
      downPred = 'y'
    if y_test > dataset3[-2]:
      upActual = 'x'
    if y_test < dataset3[-2]:
      downActual = 'y'
    if upPred == upActual:
      print('Correct price is going UP')
      predCorrect = 1
      pass
    if downPred == downActual:
      print('Correct price is going DOWN')
      predCorrect = 2
      
      pass
    if upPred != upActual and downPred !=downActual:
      print('Wrong Prediction')
      predCorrect = 3
      pass
    print('SCALED DATA last record', scaled_data[-1])
    print('SCALED DATA 4 from the last record', scaled_data[-4])
    print('SCALED DATA',scaled_data)
    print('DATASET LAST RECORD', dataset3[-1])
    RESULTS.append(predCorrect)
    print(RESULTS)
    ones=  RESULTS.count(1)
    twos = RESULTS.count(2)
    zeros = RESULTS.count(3)
    print(ones)
    print(twos)
    print(zeros)
    if i > 0 and zeros > 0:
      RATIO = (ones + twos)/ zeros
    
      print('Result RATIO',RATIO)
      totalCorrect = twos + zeros
      total = totalCorrect + zeros
   
      percentage = (100/total) * totalCorrect
    
      print('/////////////////////////////////////////')
      print('PERCENAGE', percentage)
    
```
